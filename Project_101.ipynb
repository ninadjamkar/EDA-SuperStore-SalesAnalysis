```python
#Project 101
import pandas as pd
import os
```

```python
#Merging 12 month data into a single CSV file
files = ['Sales_January_2019.csv', 'Sales_February_2019.csv', 'Sales_March_2019.csv', 'Sales_April_2019.csv', 'Sales_May_2019.csv', 'Sales_June_2019.csv', 'Sales_July_2019.csv', 'Sales_August_2019.csv', 'Sales_September_2019.csv', 'Sales_October_2019.csv', 'Sales_November_2019.csv', 'Sales_December_2019.csv']
df = pd.DataFrame()
for file in files:
    data = pd.read_csv(file)
    df = pd.concat([df, data], axis=0)
df.to_csv('all_data.csv', index=False)
```

```python
all_data = pd.read_csv("all_data.csv")
all_data
```

```python
#Adding month column
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data
```

```python
#cleaning the NaN rows
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df
```

```python
#cleaning the NaN rows
all_data = all_data.dropna(how='all')
all_data
```

```python
#clearing the Or rows
temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
temp_df
```

```python
#clearing the Or rows
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data
```

```python
#converting data type of month into int
all_data['Month'] = all_data['Month'].astype('int32')
all_data
```

```python
#Adding sales column
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])

all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data
```

```python
#Best month for sales
results = all_data.groupby('Month').sum()
results
```

```python
#Bar graph for determining which month has highest number of sales
import matplotlib.pyplot as plt

months = range(1,13)
plt.bar(months, results['Sales']);
plt.title('Monthly Sales')
plt.xlabel('Month number')
plt.ylabel('Sales in USD $ (x1000000)')
plt.xticks(months)
plt.show();
```

```python
#Adding a city column

def get_city(address):
  return address.split(',')[1]

def get_state(address):
  return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x)+' ('+get_state(x)+')')
all_data
```

```python
city_result = all_data.groupby('City').sum()
city_result
```

```python
#Determining which city has maximum sale
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, city_result['Sales'])
plt.title('Sales by each city')
plt.xlabel('City')
plt.ylabel('Sales in USD $ (x1000000)')
plt.xticks(cities, rotation='vertical', size=8)
plt.show();
```

```python
#converting order date into datetime
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Count'] = 1
all_data
```

```python
#Best time for advertisement
hours = [hour for hour, df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby(['Hour']).count());
plt.xticks(hours);
plt.xlabel('Hour');
plt.ylabel('Number of orders');
plt.grid();
```

```python
#Best products sold in combination
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

keys = [pair for pair, df in product_group]
plt.bar(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
